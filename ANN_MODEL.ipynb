```python
#IMPORTING THE LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

#PREPROCESSSING LIBRARIES
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#DATA IMPORT
df = pd.read_csv('/content/Churn_Modelling.csv')
print(f'the data has {df.shape[0]} rows and {df.shape[1]} columns')
```

```python
#misisng values check
#df.isnull().sum()                  #count of missing in each attribute
round(df.isnull().mean()*100,2)  #percentage of missing value in each attribute
```

```python
# spilt the data into independent and dependent
x = df.iloc[:, 3: -1]
y = df.iloc[:, 13]
```

```python
x
```

```python
y
```

```python
# one hot enconding
x = pd.get_dummies(x, drop_first=True,dtype="int")
x.head()
```

```python
#train test split of the data
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state=9999)
```

```python
#data scaling - normalization
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Build ANN Architecture

#1. Initalize model as sequential (ANN)

#2. add layers (input, hidden, output) along with activation functions, weight istializee

#3. metrics, loss, optimizer
```

```python
# Importing required libraries for ANN

import keras
from keras.models import Sequential #ANN
from keras.layers import Dense
```

```python
#Building the ANN Acrchitecture

#FEED FORWARD MECHANISM #(Forward Propagation)

#intialize the model
model = Sequential()

# Building the Layers

#input Layer
model.add(Dense(units=11, kernel_initializer ='he_uniform',input_dim=11))
#unit: no.of.neurons

#first hidden Layer
model.add(Dense(units=8, activation ='relu', kernel_initializer='he_uniform'))

#second hidden Layer
model.add(Dense(units=6, activation ='relu', kernel_initializer ='he_uniform'))

#output Layer
model.add(Dense(units=1, activation='sigmoid'))

#sigmoid classification problem
# BACK PROPAGATION
model.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy']);
```

```python
#ANN Summary
model.summary()
```

```python
#FITTING THE DATA INSIDE THE NETWORK ARCHITECTURE
model_history = model.fit(x_test_scaled, y_test, validation_split=0.3, batch_size=10, epochs=50)
```
